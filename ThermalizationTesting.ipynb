```python
import numpy as np
import Ising
import matplotlib.pyplot as plt

N = 10
```

```python
# Thermalization test (using single flip method)
hotMap = Ising.initializeHot(N)
coldMap = Ising.initializeCold(N)

numSteps = 10000
h = 0
temp = 3

hotMapEnergy = np.zeros(numSteps)
coldMapEnergy = np.zeros(numSteps)
hotMapMagnetization = np.zeros(numSteps)
coldMapMagnetization = np.zeros(numSteps)

for i in range(numSteps):
    hotMapEnergy[i] = Ising.energyDensity(hotMap, h)
    coldMapEnergy[i] = Ising.energyDensity(coldMap, h)
    hotMapMagnetization[i] = np.sum(hotMap)
    coldMapMagnetization[i] = np.sum(coldMap)
    
    Ising.singleSiteFlip(hotMap, temp, h)
    Ising.singleSiteFlip(coldMap, temp, h)
    
    
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

xMaxPlot = 2000

ax[0].plot(hotMapEnergy, '-', label="Hot start")
ax[0].plot(coldMapEnergy, '--', label="Cold start")

ax[0].plot([0,xMaxPlot],[sum(hotMapEnergy)/numSteps]*2,'k-')
ax[0].plot([0,xMaxPlot],[sum(coldMapEnergy)/numSteps]*2,'k--')

ax[0].set_xlim([0, xMaxPlot])
ax[0].set_xlabel('MC Step')
ax[0].set_ylabel('Energy Density')
ax[0].legend()

ax[1].plot(hotMapMagnetization, '-', label="Hot start")
ax[1].plot(coldMapMagnetization, '--', label="Cold start")

ax[1].plot([0,xMaxPlot],[sum(hotMapMagnetization)/numSteps]*2,'k-')
ax[1].plot([0,xMaxPlot],[sum(coldMapMagnetization)/numSteps]*2,'k--')

ax[1].set_xlim([0, xMaxPlot])
ax[1].set_xlabel('MC Step')
ax[1].set_ylabel('Magnetization')
ax[1].legend()

fig.suptitle(f'MC Evolution Using Site Flip Method (T = {temp}, h = {h})')
#fig.tight_layout()

"""
Comments:

Thermalization takes quite a bit longer than in the 1D model, which we should expect since there are so many sites
In the 1D, we say at around 100 steps the hot and cold starts converged, while here it is more around 250-400

At higher temperatures, both the hot and cold starts converge much faster
Energy density goes to somewhere in the range [-.2, -.5] and magnetization converges on 0

With a nonzero h and reasonable low temperature (<3) the magnetization converges to entire one spin type
 (as expected) and the energy to about -3 regardless of the sign of h
"""
plt.savefig('images/siteflip.png')
plt.show()
```

```python
# Thermalization test (using cluster flip method)
hotMap = Ising.initializeHot(N)
coldMap = Ising.initializeCold(N)

numSteps = 500
h = 0
temp = 2/np.log(1 + np.sqrt(2))

hotMapEnergy = np.zeros(numSteps)
coldMapEnergy = np.zeros(numSteps)
hotMapMagnetization = np.zeros(numSteps)
coldMapMagnetization = np.zeros(numSteps)

for i in range(numSteps):
    hotMapEnergy[i] = Ising.energyDensity(hotMap, h)
    coldMapEnergy[i] = Ising.energyDensity(coldMap, h)
    hotMapMagnetization[i] = np.sum(hotMap)
    coldMapMagnetization[i] = np.sum(coldMap)
    
    Ising.singleClusterFlip(hotMap, 1/temp)
    Ising.singleClusterFlip(coldMap, 1/temp)
    
    
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

xMaxPlot = 100

ax[0].plot(hotMapEnergy, '-', label="Hot start")
ax[0].plot(coldMapEnergy, '--', label="Cold start")

ax[0].plot([0,xMaxPlot], [sum(hotMapEnergy)/numSteps]*2,'k-')
ax[0].plot([0,xMaxPlot], [sum(coldMapEnergy)/numSteps]*2,'k--')

ax[0].set_xlim([0, xMaxPlot])
ax[0].set_xlabel('MC Step')
ax[0].set_ylabel('Energy Density')
ax[0].legend()

ax[1].plot(hotMapMagnetization, '-', label="Hot start")
ax[1].plot(coldMapMagnetization, '--', label="Cold start")

ax[1].plot([0,xMaxPlot],[sum(hotMapMagnetization)/numSteps]*2,'k-')
ax[1].plot([0,xMaxPlot],[sum(coldMapMagnetization)/numSteps]*2,'k--')

ax[1].set_xlim([0, xMaxPlot])
ax[1].set_xlabel('MC Step')
ax[1].set_ylabel('Magnetization')
ax[1].legend()

fig.suptitle(f'MC Evolution Using Cluster Flip Method (T = {temp}, h = {h})')
#fig.tight_layout()

"""
Comments:

This method is definitely much faster than the previous one, since it thermalizes after only 50 steps or so
(compared to 1500). The energy density does have the opposite sign, but that's not too big of a deal (I don't think)
"""
plt.savefig('images/clusterflip.png')
plt.show()
```
