```python
import numpy as np
import Ising
import Simulations
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 100
N = 10
```

```python
def isingEval(N, temp, thermalizationSteps=50, simulationSteps=100, skipSweeps=3, bootstrapM=500):
    """
    N should be the system length

    temp is the temperature of the model

    thermizationSteps is the number of steps to evaluate before
    calculating the susceptability. This is an estimate based on the
    results shown in the ThermalizationTest.ipynb file. Since the cluster
    flip method is rather fast, we can increase this without causing too
    much of a slowdown

    simulationSteps is the number of steps to use to calculate the susceptability
    """

    # Where we will store the magnetizations
    magnetizationArr = np.zeros(simulationSteps)
    energyArr = np.zeros(simulationSteps)
    
    # This isn't really a variable since the Wolff method requires h = 0,
    # but I'll define it to make the formulas be more verbose
    # That being said ****DO NOT CHANGE THIS****
    h = 0

    # Generate our initial conditions
    # Use hot start, but it shouldn't matter since we thermalize
    isingMap = Ising.initializeHot(N)

    # Thermalize
    for i in range(thermalizationSteps):
        Ising.singleClusterFlip(isingMap, 1/temp) 

    # Now we'll actually record data
    for i in range(simulationSteps):
        magnetizationArr[i] = Ising.magnetization(isingMap)
        energyArr[i] = Ising.energyDensity(isingMap, h)
        
        Ising.singleClusterFlip(isingMap, 1/temp)
        # Now skip; not sure why we skip and flip once beforehand
        # but this was what we did in class so I'll leave it for now
        for j in range(skipSweeps):
            Ising.singleClusterFlip(isingMap, 1/temp)


    E = Ising.statisticalBootstrap(energyArr, np.mean, bootstrapM)
    varE = Ising.statisticalBootstrap(energyArr, np.var, bootstrapM)
    
    M = Ising.statisticalBootstrap(magnetizationArr, np.mean, bootstrapM)
    varM = Ising.statisticalBootstrap(magnetizationArr, np.var, bootstrapM)

    return [E, varE, M, varM]
```

```python
# Constant H
#TArr = np.linspace(1e-1, 10, 50)
#betaArr = 1/TArr

betaArr = np.linspace(1e-3, 1, 50)
TArr = 1/betaArr

criticalTemp = 2/np.log(1 + np.sqrt(2))
criticalBeta = 1/criticalTemp
tArr = (TArr - criticalTemp) / criticalTemp

fig, ax = plt.subplots(1, 4, figsize=(20,4))
    
MArr = np.zeros(len(TArr))
varMArr = np.zeros(len(TArr))
EArr = np.zeros(len(TArr))
varEArr = np.zeros(len(TArr))

for j in range(len(TArr)):
    (EArr[j], varEArr[j], MArr[j], varMArr[j]) = isingEval(N, TArr[j])

ax[0].plot(betaArr, EArr, 'o')
ax[1].plot(betaArr, varEArr, 'o')
ax[2].plot(betaArr, MArr, 'o')
ax[3].plot(betaArr, varMArr, 'o')

for i in range(4):
    ax[i].axvline(criticalBeta, linestyle='--', color='k')

ax[0].set_title("Energy")
ax[0].set_xlabel(r"$\beta$")
ax[0].set_ylabel(r"$\langle E \rangle$")

ax[1].set_title("Specific Heat")
ax[1].set_xlabel(r"$\beta$")
ax[1].set_ylabel(r"$\langle E^2 \rangle - \langle E \rangle^2$")

ax[2].set_title("Magnetization")
ax[2].set_xlabel(r"$\beta$")
ax[2].set_ylabel(r"$\langle M \rangle$")

ax[3].set_title("Susceptability")
ax[3].set_xlabel(r"$\beta$")
ax[3].set_ylabel(r"$\langle M^2 \rangle - \langle M \rangle^2$")

fig.tight_layout()
#plt.savefig('images/E&MTesting.png')
plt.show()
```

```python
NArr = [6, 8, 10, 12, 14, 16, 18, 20]
betaArr = np.linspace(1e-3, 1, 40)

TArr = 1/betaArr
criticalTemp = 2/np.log(1 + np.sqrt(2))
criticalBeta = 1/criticalTemp
tArr = (TArr - criticalTemp) / criticalTemp

chiArr = np.zeros(len(NArr))
cArr = np.zeros(len(NArr))
tMaxCArr = np.zeros(len(NArr))
tMaxChiArr = np.zeros(len(NArr))

for i in range(len(NArr)):
    fig, ax = plt.subplots(1, 2)
    
    tempChiArr = np.zeros(len(betaArr))
    tempCArr = np.zeros(len(betaArr))
    
    for j in range(len(betaArr)):
        tempCArr[j], tempChiArr[j] = Simulations.calculateObservables(NArr[i], 1/betaArr[j], simulationSteps=100, skipSweeps=3)
        
    ax[0].plot(betaArr, tempChiArr)
    ax[0].set_xlabel(r'$\beta$')
    ax[0].set_ylabel(r'$\chi$')
    
    ax[1].plot(betaArr, tempCArr)
    ax[1].set_xlabel(r'$\beta$')
    ax[1].set_ylabel(r'$C$')
    
    fig.suptitle(f'N = {NArr[i]}')
    fig.tight_layout()
    plt.show()
    
    # Now we determine the max of the graph
    cArr[i] = max(tempCArr)
    tMaxCArr[i] = tArr[np.where(tempCArr == cArr[i])]
    chiArr[i] = max(tempChiArr)
    tMaxChiArr[i] = tArr[np.where(tempChiArr == chiArr[i])]
    
    print(f'Finished N={NArr[i]}')
```

```python
from scipy.stats import linregress

# We want to fit ln(t_max) (for chi) vs ln(N)
plt.plot(NArr, abs(tMaxCArr))
plt.xlabel(r'$N$')
plt.ylabel(r'$t_{max}$')
plt.xscale('log')
plt.yscale('log')

slope, intercept, r, p, a = linregress(np.log(NArr), np.log(tMaxChiArr))
nu = -1/slope
print(nu)

slope, intercept, r, p, a = linregress(np.log(NArr), np.log(abs(tMaxCArr)))
alpha = slope
print(alpha)

slope, intercept, r, p, a = linregress(-nu * np.log(tMaxChiArr), 1/nu * np.log(NArr))
print(slope)
```
